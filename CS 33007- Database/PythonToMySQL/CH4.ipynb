### Referential Integrety ( Foreign Key Constraint)

```python
#- observe records of instructor from history department
df = pd.read_sql_query(
    "select * from instructor",conn)
df.head(15)
```

```python
#- NOw delete the History department from the department table, for referential integrety, department 
#- field of corresponding instructor would be set to NULL 
conn.cursor().execute("Delete from department where dept_name = \'History\';");
```

```python
#- observe the department table- the Hstory department is not there anymore.
df = pd.read_sql_query(
    "select * from department",conn)
df
```

```python
#- observe the instrutor table again
df = pd.read_sql_query(
    "select * from instructor",conn)
#- So far, the changes are not done in physical table, rather in intermediate result set pointed by a cursor
#- Run either Rollback or Commit
conn.rollback() #- this statement will rollback all changes
#- conn.commit() #- Finalize changes to Physical table.

df.head(15)

```
